連立方程式を解くプログラム

```python
c=[]
n=[]
q=[]
r=[]
n=[]
i=0
print("| a1 a2 | ×　| x |  ≡ | a5 |")
print("| a3 a4 | ×  | y |     | a6 |(mod N)")

print("の連立方程式を求めるプログラム")

a1=int(input("a1="))#行列の入力
a2=int(input("a2="))
a3=int(input("a3="))
a4=int(input("a4="))
a5=int(input("a5="))
a6=int(input("a6="))
N=int(input("N="))

print("\nまずは")
print("|",a1,a2,"|")
print("|",a3,a4,"| mod(",N,")")
print("の逆行列を求める\n")

c1=a1*a4-a2*a3 #行列式を解く
c.append(c1 % N)  #最小正剰余

print(c[0],"x≡1(mod",N,")")
print(c[0],"x-",N,"k=1")
print(c[0],"x+",N,"y=1として解く(y=-k)\n")

q.append (N // c[i])  #商を求める＆リストに追加
r.append (N % c[i])   #余りを求める＆リストに追加
print("(",i,")",N,"=",c[i],"*",q[i],"+",r[i])   #ユークリッドの互除法の出力①
n.append(N)

while r[i] != 0:
  i += 1
  n.append(c[i-1])
  c.append(r[i-1])
  q.append(n[i] // c[i])
  r.append(n[i] % c[i])
  
  print("(",i,")",n[i],"=",c[i],"*",q[i],"+",r[i])  #ユークリッドの互除法の出力②

  j = i
print()
print("変形してあらわす")   #余り＝　の形に変形＆逆順に並べる

while j != 0:  
  j -= 1
  print("(",j,")",r[j],"=",n[j],"-",c[j],"*",q[j])
print()

x = [""]*i       #一次不定方程式を解く
y = [""]*i

i -= 1
x[i] = 1
y[i] = -q[i]

print("一次不定方程式について解く")
print("1=(",x[i],")*",n[i],"+","(",y[i],")*",c[i])    #一次不定式の出力①
  
while i != 0:
  i -= 1
  x[i] = y[i+1]
  y[i] = (1-x[i]*n[i])//c[i]
  print(" =(",x[i],")*",n[i],"+","(",y[i],")*",c[i])  #一次不定式の出力②

print("x≡",y[i],"(mod",n[0],")")

print("よって、A^-1は")
print("A^-1 ≡  ",y[i]," × |",a4,-a2,"|")#逆行列①
print("                |",-a3,a1,"|")
b1=a4*y[i]
b2=-a2*y[i]
b3=-a3*y[i]
b4=a1*y[i]
print("     ≡        |",b1,b2,"|")#逆行列②
print("               |",b3,b4,"|")
c1=b1%N
c2=b2%N
c3=b3%N
c4=b4%N
print("     ≡        |",c1,c2,"|")#逆行列③
print("               |",c3,c4,"|(mod",N,")\n")

print("連立方程式は")
print("|",b1,b2,"|  ×　|",a1,a2,"| ×　| x | ≡ |",b1,b2,"| × |",a5,"|")#連立方程式①
print("|",b3,b4,"|    　|",a3,a4,"|     | y |    |",b3,b4,"|    |",a6,"|")
print("これより")
print("| x | ≡ |",c1,c2,"| × |",a5,"|")#連立方程式②
print("| y |    |",c3,c4,"|    |",a6,"|")
print("となる\n")
print("よって答えは")
d1=(b1*a5+b2*a6)%N
d2=(b3*a5+b4*a6)%N

print("| x | ≡ |",d1,"|")#答え
print("| y |    |",d2,"|(mod ",N,")")

```
